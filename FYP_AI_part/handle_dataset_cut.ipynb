```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')


directory = 'Colab Notebooks/FYP/data/'  # Change this to your path

# Step 2: Define the directory containing your main zipped file
main_zip_file_path = '/content/drive/MyDrive/Colab Notebooks/FYP/data/zipped/hairpin-traces-main.zip'  # Change this to your main zip file path
unzipped_directory = '/content/drive/MyDrive/Colab Notebooks/FYP/data/alotofzipped/' # Change this to your desired output path for individual zipped files

import os
import zipfile

# Step 3: Unzip the main zipped file
with zipfile.ZipFile(main_zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(unzipped_directory)

print("Main zip file has been unzipped.")

# Step 4: Unzip each individual zipped file in the directory
for filename in os.listdir(unzipped_directory):
    if filename.endswith('.zip'):
        individual_zip_file_path = os.path.join(unzipped_directory, filename)
        with zipfile.ZipFile(individual_zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(unzipped_directory)

print("All individual zip files have been unzipped.")
```

```python
unzipped_directory2 = '/content/drive/MyDrive/Colab Notebooks/FYP/data/alotofzipped/hairpin-traces-main'

import os
import zipfile
import gzip
import shutil
# Step 5: Decompress each .gz file in the directory
for filename in os.listdir(unzipped_directory2):
    if filename.endswith('.gz'):
        gz_file_path = os.path.join(unzipped_directory2, filename)
        output_file_path = os.path.join(unzipped_directory2, filename[:-3])  # Remove .gz extension

        # Decompress the .gz file
        with gzip.open(gz_file_path, 'rb') as f_in:
            with open(output_file_path, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)

print("All .gz files have been decompressed.")
```

```python

# Step 5: Process each text file in the unzipped directory
for filename in os.listdir(unzipped_directory2):
    if filename.endswith('.txt'):
        file_path = os.path.join(unzipped_directory2, filename)

        # Read the file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Count the number of rows
        num_rows = len(lines)

        # Calculate the number of rows to keep (largest multiple of 10,000)
        rows_to_keep = num_rows - (num_rows % 10000)

        # Trim the lines to keep
        trimmed_lines = lines[:rows_to_keep]

        # Write the trimmed lines back to the file
        with open(file_path, 'w') as file:
            file.writelines(trimmed_lines)

print("All text files have been processed and trimmed.")
```

```python
unzipped_directory3 = '/content/drive/MyDrive/Colab Notebooks/FYP/data/'
# Step 2: Define the directory containing your .log files
log_directory = unzipped_directory2  # Change this to your log files path
output_excel_file = '/content/drive/MyDrive/Colab Notebooks/FYP/data/output_file.xlsx'  # Change this to your desired output path

import pandas as pd
import os

# Step 3: Initialize an empty list to hold all the data
all_data = []

# Step 4: Read each .log file and extract relevant data
for filename in os.listdir(log_directory):
    if filename.endswith('.log'):
        file_path = os.path.join(log_directory, filename)

        # Read the log file
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Process each line and extract relevant data
        for line in lines:
            # Assuming each line has three values separated by whitespace
            # Modify the split logic according to your actual data format
            values = line.strip().split()  # Split by whitespace
            if len(values) >= 3:  # Ensure there are at least 3 values
                # Append the first three values to the all_data list
                all_data.append(values[:3])

# Step 5: Create a DataFrame from the combined data
df = pd.DataFrame(all_data, columns=['bandwidth', 'delay', 'packetloss'])

# Step 6: Save the DataFrame to an Excel file
df.to_excel(output_excel_file, index=False)

print("All .log files have been concatenated and saved to Excel.")
```
